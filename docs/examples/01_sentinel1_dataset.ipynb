```python
import ee
import geemap

from geersd import Sentinel1
```

```python
ee.Initialize()
```

```python
AOI = ee.Geometry.Point([-77.39244297585545, 44.169272034699034])
DATE = ("2019-04-01", "2019-10-31")
```

```python
s1_obj = (
    Sentinel1()
    .filterBounds(AOI)
    .filterDate(*DATE)
    .filterIWMode()
    .filterVV()
    .filterVH()
    .applyEdgeMask()
)
```

```python
s1_obj
```

```python
s1_obj.aggregate_array("instrumentMode").distinct()
```

```python
s1_obj.aggregate_array("transmitterReceiverPolarisation").distinct()
```

```python
Map = geemap.Map()

Map.addLayer(s1_obj.median(), {"bands": ["VV"], "min": -25, "max": 10})

Map
```

```python
s1_asc = s1_obj.filterAsc()
```

```python
s1_asc.aggregate_array("orbitProperties_pass").distinct()
```

```python
s1_desc = s1_obj.filterDesc()
```

```python
s1_desc.aggregate_array("orbitProperties_pass").distinct()
```
